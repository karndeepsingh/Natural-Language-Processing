```python
import nltk
from nltk.corpus import stopwords
import re
from nltk.stem import WordNetLemmatizer
wl = WordNetLemmatizer()
corpus = []
for i in range(len(tweets["text"])):
    
    text = re.sub("[^a-zA-z]", " ", tweets["text"][i])
    text = text.lower()
    text = text.split()
    text = [wl.lemmatize(word) for word in text if word not in set(stopwords.words("english"))]
    corpus.append(" ".join(text))
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features = 6000)
txt = cv.fit_transform(corpus).toarray()
```

```python
strings = " ".join(corpus) 
```

```python
from wordcloud import WordCloud
```

```python
word_cloud = WordCloud(width = 1200, height = 600, margin = 2).generate(strings)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (10,10))
plt.imshow(word_cloud,interpolation = "gaussian")
plt.axis("off")
plt.show()
```
